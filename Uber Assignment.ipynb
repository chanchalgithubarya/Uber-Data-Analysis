# Q15. What is the most popular starting point for the Uber drivers?
## Note: This question is based on dataframe with no 'na' values in the 'START' variable

```python
starting_point = df['START*'].dropna()
pop_start_point = pd.DataFrame(starting_point.value_counts()).sort_values('START*',ascending=False).reset_index()
pop_start_point.rename(columns={'index':'Starting Point','START*':'Count'}).head(1)
```

# Q16. What is the most popular dropping point for the Uber drivers?
## Note: This question is based on dataframe with no 'na' values in the 'STOP*' variable

```python
stop_point = df['STOP*'].dropna()
pop_stop_point= pd.DataFrame(stop_point.value_counts()).sort_values('STOP*',ascending=False).reset_index().rename(columns={'index':'Dropping Point','STOP*':'Count'})

pop_stop_point.loc[pop_stop_point['Count'] == max(pop_stop_point['Count'])]
```

# Q17. List the most frequent route taken by Uber drivers.
## Note: This question is based on dataframe with no 'na' values.

```python
df2 = df.dropna()
freq_route = pd.DataFrame(df2.groupby(['STOP*',"START*"]).size()).rename(columns={0:'Count'})

freq_route = freq_route.sort_values('Count',ascending=False).head(1)
freq_route
```

# Q22. What is proportion of trips that is Business and what is the proportion of trips that is Personal?
## Note: This proportion calculation is with respect to 'Miles' variable.
## Note: Based on original dataframe

```python

df1 = uber.groupby('CATEGORY*').sum()
df1
```

```python
Business = df1.iloc[0,0]/(df1.iloc[0,0]+df1.iloc[1,0])
Personal = df1.iloc[1,0]/(df1.iloc[0,0]+df1.iloc[1,0])

print('Business:',Business)
print('Personal:',Personal)
```
